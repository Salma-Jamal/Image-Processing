**Imports**

```python
import cv2
from google.colab.patches import cv2_imshow
```

**Read Image**

```python
img = cv2.imread("/content/img.jpg")
```

```python
cv2_imshow(img)
```

**Detection of the Edges**

```python
detect_edge = cv2.Canny(img,100,200)
```

```python
cv2_imshow(detect_edge)
```

**Final Output**

```python
for i in range(len(detect_edge)):
  for j in range(len(detect_edge[0])):
    if detect_edge[i][j] == 255:
      img[i][j] = detect_edge[i][j]
cv2_imshow(img)
```
